```python
import matplotlib.pyplot as plt
import numpy as np

# Set LaTeX font parameters
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "axes.labelsize": 15,
    "font.size": 20,
    "legend.fontsize": 15,
    "xtick.labelsize": 15,
    "ytick.labelsize": 15,
    "figure.titlesize": 16,
    "figure.dpi": 300,
})

# Data
widths = [16, 32, 64]
bfno_psnr = [53.7, 56.34, 56.86]
fno_psnr = [49.36, 51.0, 52.44]

# Setting up the figure and axis
plt.figure(figsize=(10, 6))

# Creating the line plots with markers
plt.plot(widths, bfno_psnr, 'o-', color='#8884d8', linewidth=2, markersize=8, label=r'Bicubic FNO')
plt.plot(widths, fno_psnr, 's-', color='#82ca9d', linewidth=2, markersize=8, label=r'FNO')

# Adding data point annotations
for i, (x, y) in enumerate(zip(widths, bfno_psnr)):
    plt.annotate(f"${y}$", (x, y), textcoords="offset points", 
                xytext=(0, 10.5), ha='center', fontsize=20)

for i, (x, y) in enumerate(zip(widths, fno_psnr)):
    plt.annotate(f"${y}$", (x, y), textcoords="offset points", 
                xytext=(0, 10.5), ha='center', fontsize=20)

# Adding labels and title
plt.xlabel(r'Width', fontsize=25)
plt.ylabel(r'PSNR (dB)', fontsize=25)
# plt.title(r'FNO vs Bicubic FNO Comparison', fontsize=30, fontweight='bold')

# Setting x-axis ticks
plt.xticks(widths, fontsize=25)
# Setting y-axis ticks
plt.yticks(np.arange(45, 65, 5), fontsize=25)

# Setting x and y limits
plt.xlim(10, 70)
plt.ylim(40, 60)

# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Customizing tick parameters
plt.tick_params(axis='both', which='major', direction='out', length=4, width=1)

# Adding a legend with a frame
legend = plt.legend(frameon=True, loc='lower right')
legend.get_frame().set_linewidth(0.8)

plt.tight_layout()

plt.show()

plt.savefig('bfno_vs_fno_lineplot.png', bbox_inches='tight', dpi = 300)
```

```python
# Set LaTeX font parameters
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "axes.labelsize": 11,
    "font.size": 10,
    "legend.fontsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "figure.titlesize": 14,
    "figure.dpi": 300,
})

# Data
widths = [16, 32, 64]
bfno_psnr = [53.7, 56.34, 56.86]
fno_psnr = [49.36, 51.0, 52.44]

# Setting up the figure and axis
plt.figure(figsize=(10, 6))

# Bar width
width = 0.35

# X positions for bars
x = np.arange(len(widths))

# Creating the bar plots
plt.bar(x - width/2, fno_psnr, width, label='FNO', color='#82ca9d')
plt.bar(x + width/2, bfno_psnr, width, label='Bicubic FNO', color='#8884d8')

# Adding labels and title
plt.xlabel(r'Width', fontsize=11)
plt.ylabel(r'PSNR (dB)', fontsize=11)
plt.title(r'FNO vs Bicubic FNO Comparison', fontsize=14, fontweight='bold')

# Setting the x-ticks to match our width values
plt.xticks(x, widths)

# Setting y-axis limits 
plt.ylim(45, 60)

# Adding value labels on top of each bar
for i, v in enumerate(fno_psnr):
    plt.text(i - width/2, v + 0.3, f"${v}$", ha='center', fontsize=9)
    
for i, v in enumerate(bfno_psnr):
    plt.text(i + width/2, v + 0.3, f"${v}$", ha='center', fontsize=9)

# Adding a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding a legend
plt.legend()

plt.tight_layout()

plt.show()

# You can also save the figure with:
plt.savefig('bfno_vs_fno_comparison.png', dpi=300, bbox_inches='tight')
```
